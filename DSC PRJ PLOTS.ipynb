```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
df=pd.read_csv("Breast-Cancer-Dataset.csv")
```

```python
df
```

```python
df.groupby('diagnosis(M=1, B=0)').size()
```

```python
sns.displot(df['diagnosis(M=1, B=0)'])
plt.show()
#To show number of records for each diagnosis (1 and 0) 
```

```python
plt.plot(df['radius_mean'],df['diagnosis(M=1, B=0)'])
#plt.boxplot(df['area_mean'])
#plt.show()
```

```python
plt.scatter(df['radius_mean'], df['diagnosis(M=1, B=0)'])
plt.show()
```

```python
breast_cancer = df.drop('id', axis=1)
 
breast_cancer_corr = breast_cancer.corr()
 
print(breast_cancer_corr)
```

```python
plt.figure(figsize=(8, 8))
 
sns.heatmap(breast_cancer_corr, cbar=True, annot=False, yticklabels=breast_cancer.columns,
            xticklabels=breast_cancer.columns)
plt.show()
```

```python
high_correlation = breast_cancer_corr.abs()
high_correlation_unstack=high_correlation.unstack()
high_correlation_sort = high_correlation_unstack.sort_values(ascending=False)
 
print(high_correlation_sort[32:37])#must NOT include 1.000000 
```

```python
sns.jointplot("radius_worst", "perimeter_worst", data=breast_cancer, kind="scatter",
              space=0, color="#FF1284", height=5, ratio=3)
plt.show()
```

```python
sns.pairplot(breast_cancer, vars=["radius_worst", "perimeter_worst"],
             palette=sns.color_palette(['#FF1493', '#FF69B4']), hue='diagnosis(M=1, B=0)', height=3)
plt.show()
```
